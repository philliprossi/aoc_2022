```python
import numpy as np

class FooWrapper(object):
    def __init__(self, value):
         self.value = value

    def __str__(self):
        return (str(self.value))

with open('inputs/day20.txt') as f:
    original_list = f.readlines()
    original_list = [int(x) for x in original_list]
    original_list = [FooWrapper(x) for x in original_list]

    original_values = np.array(original_list)
    adjusted_values = np.array(original_list)

    for item in original_values:

        print('----HERE-----')
        #print(f'itemn: {item}')
        #for x in adjusted_values:
        #    print(x, end=', ')
        #print('')


        element_location = np.where(adjusted_values == item)[0][0]

        print(f'element_location: {element_location}')

        print(f'value to move: {adjusted_values[element_location]}')

        moves = adjusted_values[element_location].value

        if moves >=0:
            print('HERE')
            moves_made = 0
            
            move_start = (element_location) % len(adjusted_values)
            move_start_end = (element_location+1) % len(adjusted_values)

            print(move_start, move_start_end)
            while True:
                if moves_made>=moves:
                    break

                valuetomove = adjusted_values[move_start]
                adjusted_values = np.delete(adjusted_values, move_start)
                adjusted_values = np.insert(adjusted_values, move_start_end, valuetomove)
                moves_made += 1

                move_start = move_start_end
                move_start_end = move_start_end +1

                if move_start_end == len(adjusted_values):
                    move_start_end = 1
                if move_start_end == len(adjusted_values) - 1:
                    move_start_end = 0

        else:
            moves_made = 0
            move_start = (element_location) 
            move_start_end = (element_location-1) 
            
            if move_start_end<0:
                move_start_end = len(adjusted_values) -2

            while True:
                if moves_made<=moves:
                    break

                valuetomove = adjusted_values[move_start]
                adjusted_values = np.delete(adjusted_values, move_start)
                adjusted_values = np.insert(adjusted_values, move_start_end, valuetomove)
                moves_made -= 1

                move_start = move_start_end
                move_start_end = move_start_end -1

                if move_start_end<0:
                    move_start_end = len(adjusted_values) -2
                if move_start_end == 0:
                    move_start_end = len(adjusted_values) -1


    print('----FINAL ---')
    for x in adjusted_values:
        print(x)

    print(adjusted_values)
    print(original_values)
```
