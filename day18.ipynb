```python
from builtins import max
import pprint

def check_side(point, points):
    if (point[0], point[1], point[2]) in points:
        return - 1
    else:
        return  1

def check_if_side_unbounded(point,points, outer_limits, visited):

    visited.append(point)

    if point in points:
        return visited

    if check_for_unbounded_point(visited, outer_limits): #exit early if point already found
        return visited

    #check every side
    #if any of them are unbounded check themn
    #stop when you are outside lava box
    if point[0] <= outer_limits[0] or point[0] >= outer_limits[1] or \
        point[1] <= outer_limits[2] or point[1] >= outer_limits[3] or \
        point[2] <= outer_limits[4] or point[2] >= outer_limits[5]:

        return visited

    point_xm1 = (point[0]-1, point[1], point[2])
    if point_xm1 not in points and point_xm1 not in visited:
        check_if_side_unbounded(point_xm1, points,outer_limits,visited)

    point_xp1 = (point[0]+1, point[1], point[2])
    if point_xp1 not in points and point_xp1 not in visited:
        check_if_side_unbounded(point_xp1, points,outer_limits,visited)

    point_ym1 = (point[0], point[1]-1, point[2])
    if point_ym1 not in points and point_ym1 not in visited:
        check_if_side_unbounded(point_ym1, points,outer_limits,visited)
    point_yp1 = (point[0], point[1]+1, point[2])
    if point_yp1 not in points and point_yp1 not in visited:
        check_if_side_unbounded(point_yp1, points,outer_limits,visited)

    point_zm1 = (point[0], point[1], point[2]-1)
    if point_zm1 not in points and point_zm1 not in visited:
        check_if_side_unbounded(point_zm1, points,outer_limits,visited)
    point_zp1 = (point[0], point[1], point[2]+1)
    if point_zp1 not in points and point_zp1 not in visited:
        check_if_side_unbounded(point_zp1, points,outer_limits,visited)


    return visited

def check_for_unbounded_point(points, outer_limits):
    for point in points:

        if point[0] <= outer_limits[0] or point[0] >= outer_limits[1] or \
        point[1] <= outer_limits[2] or point[1] >= outer_limits[3] or \
        point[2] <= outer_limits[4] or point[2] >= outer_limits[5]:
            return True
        
    return False


with open('inputs/day18.txt') as f:
    moves = f.read()


    points = {}
    unconnected_sides = 0
    for point in moves.split():

        point = tuple(int(num) for num in point.split(','))
        points[point] = point

    minx, max_x = 0,0
    miny, maxy = 0,0
    minz, maxz = 0,0

    for point in points:
        minx = min(minx, point[0])
        max_x = max(max_x, point[0])

        miny = min(miny, point[1])
        maxy = max(maxy, point[1])

        minz = min(minz, point[2])
        maxz = max(maxz, point[2])

    minx+= -1
    max_x+=1

    miny+= -1
    maxy+=1  

    minz+= -1
    maxz+=1   

    print(f"minx: {minx}, max_x:{max_x}, miny: {miny}, maxy: {maxy}, minz: {minz}, maxz: {maxz}")


    unbounded = 0
    for point in points:

        #point = (2, 2, 4)
        #check for every side the point right to the side of the cell
        #point is unbounded if one of the visiteed cells is outside the box
        print(f"point: {point}")
        visited = []
        check_if_side_unbounded((point[0]+1, point[1], point[2]), points, (minx, max_x, miny, maxy, minz, maxz),visited)
        if check_for_unbounded_point(visited, (minx, max_x, miny, maxy, minz, maxz)):
            unbounded+=1

        visited = []
        check_if_side_unbounded((point[0]-1, point[1], point[2]), points, (minx, max_x, miny, maxy, minz, maxz),visited)
        if check_for_unbounded_point(visited, (minx, max_x, miny, maxy, minz, maxz)):
            unbounded+=1

        visited = []
        check_if_side_unbounded((point[0], point[1]+1, point[2]), points, (minx, max_x, miny, maxy, minz, maxz),visited)
        if check_for_unbounded_point(visited, (minx, max_x, miny, maxy, minz, maxz)):
            unbounded+=1

        visited = []
        check_if_side_unbounded((point[0], point[1]-1, point[2]), points, (minx, max_x, miny, maxy, minz, maxz),visited)
        if check_for_unbounded_point(visited, (minx, max_x, miny, maxy, minz, maxz)):
            unbounded+=1

        visited = []
        check_if_side_unbounded((point[0], point[1], point[2]+1), points, (minx, max_x, miny, maxy, minz, maxz),visited)
        if check_for_unbounded_point(visited, (minx, max_x, miny, maxy, minz, maxz)):
            unbounded+=1

        visited = []
        check_if_side_unbounded((point[0], point[1], point[2]-1), points, (minx, max_x, miny, maxy, minz, maxz),visited)
        if check_for_unbounded_point(visited, (minx, max_x, miny, maxy, minz, maxz)):
            unbounded+=1

    print(unbounded)


```
